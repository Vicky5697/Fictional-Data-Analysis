### Data Visualization

### Gender Count

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data['Gender'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Count')
plt.show()
```

```python
data.Gender.value_counts()
```

### Illness Count

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data['Illness'])
plt.xlabel('Illness')
plt.ylabel('Count')
plt.title('Illness Count')
plt.show()
```

### City Count

```python
plt.figure(figsize=(15,5))
sns.countplot(x=data['City'])
plt.xlabel('City')
plt.ylabel('Count')
plt.title('City Count')
plt.show()
```

```python
data.City.value_counts()
```

### Illness by Gender

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data['Gender'], hue=data['Illness'])
plt.xlabel('Gender')
plt.ylabel('Illness Count')
plt.title('Illness count by Gender')
plt.show()
```

```python
data.Illness.value_counts()
```

### Gender Count by City

```python
plt.figure(figsize=(15,5))
sns.countplot(x=data['Gender'], hue=data['City'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Count by City')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x=data['City'], hue=data['Gender'])
plt.xlabel('City')
plt.ylabel('Count')
plt.title('City Count by Gender')
plt.show()
```

### Age Distribution

```python
plt.figure(figsize=(10,5))
sns.distplot(x=data['Age'])
plt.xlabel('Age')
plt.title('Age distribution')
plt.show()
```

### Income Distribution

```python
plt.figure(figsize=(10,5))
sns.distplot(x=data['Income'])
plt.xlabel('Income')
plt.title('Income Distribution')
plt.show()
```

### Income Distribution Illness : Yes

```python
plt.figure(figsize=(10,5))
sns.distplot(x=data[data['Illness'] == 'Yes']['Income'])
plt.xlabel('Income Distribution')
plt.title('Income Distribution - Illness : Yes')
plt.show()
```

### Income Distribution Illness : No

```python
plt.figure(figsize=(10,5))
sns.distplot(x=data[data['Illness'] == 'No']['Income'])
plt.xlabel('Income Distribution')
plt.title('Income Distribution - Illness : No')
plt.show()
```

### Income Distribution - Man and Woman

```python
plt.figure(figsize=(10,5))
sns.distplot(data[data['Gender'] == 'Male']['Income'])
sns.distplot(data[data["Gender"] == "Female"]["Income"])
plt.legend(labels=['Male','Female'])
plt.title("Income distribution - Man and Woman")
plt.show()
```

### Income by City

```python
plt.figure(figsize=(15,5))
sns.boxplot(x=data['Income'], y=data['City'])
plt.xlabel('Income')
plt.ylabel('City')
plt.title('Income By City')
plt.show()
```

### Income by Gender

```python
plt.figure(figsize=(15,5))
sns.boxplot(x=data['Income'], y=data['Gender'])
plt.xlabel('Income')
plt.ylabel('Gender')
plt.title('Income By Gender')
plt.show()
```

### Income By Age

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=data['Age'], y=data['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income By Age')
plt.show()
```

### Comparison of Gender

```python
plt.figure(figsize=(10,5))
values = data['Gender'].value_counts().values
counts = data['Gender'].value_counts().index
colors = ['Blue','Orange']
plt.pie(values,labels = counts,colors = colors,autopct='%1.1f')
plt.title('Comparision of Gender')
plt.legend()
plt.show()
```

### Income Distribution by Gender

```python
Male = data[data['Gender'] == 'Male']
Female = data[data['Gender'] == 'Female']

x = pd.Series(Male['Income'])
y = pd.Series(Female['Income'])

plt.figure(figsize=(10,5))
plt.hist(x, label='Male')
plt.hist(y, label='Female')
plt.title('Income Distribution by Gender')
plt.xlabel('Income')
plt.ylabel('Count')
plt.legend()
plt.show()
```
